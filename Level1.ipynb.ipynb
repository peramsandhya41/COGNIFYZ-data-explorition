### Aggregate Rating Analysis

The distribution of aggregate ratings shows that a large number of restaurants have a rating of 0.0, which may indicate that these restaurants have not been rated yet.

Most of the restaurants have ratings between 3.0 and 4.0, indicating average to good performance.

Very few restaurants have extremely high or low ratings, suggesting that the data is slightly skewed towards moderate ratings.

```python
import matplotlib.pyplot as plt

df["Aggregate rating"].hist()
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Aggregate Ratings")
plt.show()
```

```python
import folium

# Create base map using average location
m = folium.Map(location=[df["Latitude"].mean(), df["Longitude"].mean()], zoom_start=5)

# Add markers (first 100 rows)
for i in range(100):
    folium.Marker(
        location=[df.iloc[i]["Latitude"], df.iloc[i]["Longitude"]],
        popup=df.iloc[i]["Restaurant Name"]
    ).add_to(m)

m
```

### Geospatial Analysis

The geographical distribution of restaurants was visualized using latitude and longitude coordinates on an interactive map.

The map shows that restaurants are highly concentrated in major cities such as New Delhi, Gurgaon, and Noida.

This indicates that urban areas have a higher density of restaurants. No strong direct relationship is observed between location and ratings, but cities with more restaurants show a wider variation in ratings.
